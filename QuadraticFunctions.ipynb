```python
# Finding the roots of a quadratic equation
import cmath

# Coefficients of the quadratic equation
a = 1
b = -5
c = 6

# Calculate the discriminant
D = (b**2) - (4*a*c)

# Find the roots
root1 = (-b + cmath.sqrt(D)) / (2*a)
root2 = (-b - cmath.sqrt(D)) / (2*a)

# Print the roots
print("Root 1:", root1)
print("Root 2:", root2)
```

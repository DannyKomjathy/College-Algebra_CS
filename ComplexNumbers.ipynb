This script defines the Arithmetic Operations of Complex Numbers.

```python
import math

# Function to add two complex numbers
def add_complex(a, b):
    real_part = a.real + b.real
    imaginary_part = a.imag + b.imag
    return complex(real_part, imaginary_part)

# Function to subtract two complex numbers
def subtract_complex(a, b):
    real_part = a.real - b.real
    imaginary_part = a.imag - b.imag
    return complex(real_part, imaginary_part)

# Function to multiply two complex numbers
def multiply_complex(a, b):
    real_part = (a.real * b.real) - (a.imag * b.imag)
    imaginary_part = (a.real * b.imag) + (a.imag * b.real)
    return complex(real_part, imaginary_part)

# Function to divide two complex numbers
def divide_complex(a, b):
    denominator = (b.real ** 2) + (b.imag ** 2)
    real_part = ((a.real * b.real) + (a.imag * b.imag)) / denominator
    imaginary_part = ((a.imag * b.real) - (a.real * b.imag)) / denominator
    return complex(real_part, imaginary_part)

# Function to compute the magnitude of a complex number
def compute_magnitude(a):
    magnitude = math.sqrt((a.real ** 2) + (a.imag ** 2))
    return magnitude

# Function to compute the conjugate of a complex number
def compute_conjugate(a):
    return complex(a.real, -a.imag)

# Main program
if __name__ == "__main__":
    # Test the complex number arithmetic operations
    complex_num1 = complex(3, 4)
    complex_num2 = complex(2, -1)

    # Addition
    sum_result = add_complex(complex_num1, complex_num2)
    print(f"Sum: {sum_result}")

    # Subtraction
    diff_result = subtract_complex(complex_num1, complex_num2)
    print(f"Difference: {diff_result}")

    # Multiplication
    product_result = multiply_complex(complex_num1, complex_num2)
    print(f"Product: {product_result}")

    # Division
    quotient_result = divide_complex(complex_num1, complex_num2)
    print(f"Quotient: {quotient_result}")

    # Magnitude
    magnitude = compute_magnitude(complex_num1)
    print(f"Magnitude of complex_num1: {magnitude}")

    # Conjugate
    conjugate = compute_conjugate(complex_num1)
    print(f"Conjugate of complex_num1: {conjugate}")
```

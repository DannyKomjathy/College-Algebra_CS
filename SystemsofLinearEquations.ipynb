```python
# Solving a system of linear equations using the Gauss-Jordan elimination method
import numpy as np

# Coefficient matrix
A = np.array([[2, 3, -1], [4, -1, 2], [1, -2, 1]])

# Constant vector
b = np.array([4, 3, -1])

# Augmented matrix
augmented_matrix = np.column_stack((A, b))

# Apply Gauss-Jordan elimination
rref_matrix = np.linalg.matrix_rank(augmented_matrix)

# Extract the solution
solution = rref_matrix[:, -1]

# Print the solution
print("Solution:", solution)
```

```python
# Calculating the sum of an arithmetic series
# using the formula: Sn = (n/2)(a + L), where n is the number of terms,
# a is the first term, and L is the last term

# Number of terms
n = 10

# First and last term
a = 2
L = 20

# Calculate the sum
sum = (n/2) * (a + L)

# Print the sum
print("Sum:", sum)
```
